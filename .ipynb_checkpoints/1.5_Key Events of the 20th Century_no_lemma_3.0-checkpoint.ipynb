# Text mining LOTR

```python
from textblob import TextBlob 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

## Tokenization

```python
myfile = open('Key events of the 20th century_without references.txt', encoding='utf-8') 
```

```python
# Import txt file

with open('Key events of the 20th century_without references.txt', 'r', errors='ignore') as file:
    data = file.read().replace('\n', ' ').replace('\t', ' ')
```

```python
# Sentence tokenization 

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent)
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
# Create frequency distribution

from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words)
```

```python
dist_words.most_common(10)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words.plot(10,cumulative = False)
plt.show()
```

## Stopwords

```python
# Defining stopwords

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
# Removing stopwords in words

filtered_words = [] # creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
filtered_words
```

```python
# Create a new FreqDist for filtered_words

dist_words_filter = FreqDist(filtered_words)
print(dist_words_filter)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter.plot(10, cumulative = False)
plt.show()
```

```python
dist_words_filter
```

```python
# Substitute all punctuations marks with a space 

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                          " ",          # Replace all non-letters with spaces
                          str(filtered_words))
```

```python
sans_punc
```

```python
# Word tokenization

tokenized_word_2 = word_tokenize(sans_punc)
print(tokenized_word_2)
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(30, cumulative = False)
plt.show()
```

```python
dist_words_filter_2.most_common(20)
```

# Group similar words

```python
word_map = {
    # wars
    'war': 'war',
    'wars': 'war',
    'attack': 'attack',
    'attacks': 'attack',
    'Second World War': 'worldwar2',
    'World War II': 'worldwar2',
    'II': 'worldwar2',
    'WWII': 'worldwar2',

    # world & alliances
    'world': 'world',
    'World': 'world',
    'Allied': 'allies',
    'allied': 'allies',

    # countries & adjectives
    'Germany': 'germany',
    'German': 'germany',
    'Germans': 'germany',
    'germany': 'germany',
    'german': 'germany',
    'germans': 'germany',

    'European': 'europe',
    'european': 'europe',

    'United States': 'us',
    'united states': 'us',
    'United States of America': 'us',
    'u.s.a.': 'us',
    'USA': 'us',
    'US': 'us',
    'America': 'us',
    'america': 'us',

    'Great Britain': 'uk',
    'great britain': 'uk',
    'England': 'uk',
    'england': 'uk',
    'British': 'uk',
    'british': 'uk'
}


new_stopwords = ["the", "u", "The", "many", "s", "This", "would", "could", "edit", "in", "new", "U", "also", "main", "Main", "th", "In", "end", "became"]
```

```python
# Create new normalized FreqDist (with lowercasing)
normalized_dist = FreqDist()

for word, freq in dist_words_filter_2.items():
    # lowercase before lookup
    w = word.lower()

    # skip stopwords
    if w in new_stopwords:
        continue

    # if the word is in the mapping, replace it; otherwise keep the lowercase form
    base = word_map.get(w, w)

    # add frequency
    normalized_dist[base] += freq
```

```python
# Show the 20 most common normalized words
for word, freq in normalized_dist.most_common(20):
    print(f"{word:15} {freq}")
```

```python
# This analysis produced the 20 most frequently used words in the text, 
# revealing the dominant historical themes and key entities mentioned throughout the 20th-century narrative.
```

###  Summary of Most Frequent Words

```python
# The most frequent words highlight a clear focus on global conflict and power dynamics of the twentieth century.
# Terms like “war”, “Germany”, “Soviet Union”, and “Japan” dominate the text, confirming that World War II 
# and its aftermath are central themes. Frequent mentions of “Allies”, “Hitler”, and “World War II” reinforce this narrative.

# Words such as “nuclear”, “weapons”, and “attack” suggest strong emphasis on military technology and the emerging nuclear age. 
# References to “Europe”, “United”, and “States” underline the international scope of the analysis, 
# showing how nations and alliances shaped global events across the century.
```

## POS Tags with TextBlob

```python
normalized_tokens = []

for word in tokenized_word_2:  # or whatever your list of tokens is called
    w = word.lower()

    # skip stopwords
    if w in new_stopwords:
        continue

    # map aliases to canonical form
    base = word_map.get(word, word_map.get(w, w))
    normalized_tokens.append(base)
```

```python
print(len(normalized_tokens)) 
```

```python
%%time
text = TextBlob(" ".join(normalized_tokens))
```

```python
text
```

```python
tags_list = text.tags
```

```python
tags_list
```

```python
bad_tokens = [
    # navigation / UI / Wikipedia boilerplate
    'key', 'events', 'century', 'wikipedia', 'jump', 'content', 'menu', 'move',
    'sidebar', 'hide', 'navigation', 'pagecontentscurrent', 'eventsrandom',
    'articleabout', 'wikipediacontact', 'contribute', 'helplearn',
    'editcommunity', 'portalrecent', 'changesupload', 'filespecial', 'pages',
    'search', 'appearance', 'donate', 'create', 'account', 'contact',

    # capitalized / alternate forms (in case any remain)
    'pagecontentscurrent', 'eventsrandom', 'articleabout', 'wikipediacontact',
    'portalrecent', 'filespecial', 'wikipedia', 'navigation', 'jump',
]    

cleaned_tags = [ (word, tag) for (word, tag) in tags_list if word not in bad_tokens ]
```

```python
df_text = pd.DataFrame(cleaned_tags)
df_text.columns = ['Words', "Word type"]
```

```python
len(cleaned_tags)
```

```python
df_text.head()
```

```python
df_t = df_text.groupby('Word type').count().reset_index()
```

```python
df_t.head()
```

```python
top10 = df_t.nlargest(10, 'Words')
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(
    x="Words",
    y="Word type",
    data=top10
)
ax.set_title("Key events of the 20th century - Top 10 word types used")
plt.show()
```

```python
df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNP")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
```

## Counting country names occurences

```python
# Mapping short country codes back to full names
country_backmap = {
    'us': 'united states',
    'uk': 'united kingdom',
    'ussr': 'russia',
    'soviet union': 'russia',
    'soviet': 'russia'
}

# Replace short codes in your normalized_tokens
normalized_tokens_back = [country_backmap.get(w.lower(), w.lower()) for w in normalized_tokens]

# Count how often each appears
from collections import Counter
all_counts = Counter(normalized_tokens_back)

print(all_counts['united states'])
print(all_counts['united kingdom'])
print(all_counts['russia'])
```

```python
# Create a count for countries

all_counts = Counter(normalized_tokens_back)
```

```python
all_counts
```

```python
countries = pd.read_csv("List of Countries_2.0.csv", index_col = 0)
```

```python
countries.head()
```

```python
# lowercase country names
countries_list = [c.lower() for c in countries['Country'].to_list()]
```

```python
dict_of_counts = {d : all_counts[d] for d in countries_list}
```

```python
dict_of_counts
```

```python
# Search for the countries from the list in the dictionary

dct = {v:[k] for v,k in dict_of_counts.items()}  
df = pd.DataFrame(dct)
```

```python
df
```

```python
df = df.transpose().reset_index()
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.rename(columns = {"index":"Country", 0:"Times mentioned"}, inplace = True)
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.groupby("Country")["Times mentioned"].sum().reset_index()
```

```python
df_top10 = df.sort_values("Times mentioned", ascending=False).head(10)
```

```python
df_top10.head(10)
```

```python
# Plot
plt.figure(figsize=(10, 5))
ax = sns.barplot(
    x="Times mentioned",
    y="Country",
    data=df_top10
)

# Add title
ax.set_title("Most frequently mentioned countries")
plt.show()
```

The analysis of twentieth-century events shows that Germany dominates the narrative, being mentioned far more often than any other country — a reflection of its central role in both World Wars and post-war European reconstruction.
Japan and the United Kingdom follow, highlighting their significance in global conflicts and post-war politics.
Mentions of France, Poland, and the United States also feature prominently, pointing to their involvement in major geopolitical shifts of the century.
Meanwhile, countries such as China, Italy, Vietnam, and Russia appear less frequently but still mark crucial regional turning points and ideological confrontations during the Cold War era.
